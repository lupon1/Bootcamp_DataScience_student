```python
import pandas as pd
import numpy as np
csv = 'C:\\Users\\lupon\\Dropbox\\ID_BOOTCAMPS\\Bootcamp_DataScience_student\\week10\\data\\parsed.csv'
data = pd.read_csv(csv)
data.head()
```

```python
# - Encuentre el percentil 95 de la magnitud del terremoto en Japón usando el tipo de magnitud `mb`
data1 = data[(data.magType == 'mb') & (data.parsed_place == 'Japan')]
mags = list(data1['mag'])
perc = np.percentile(mags, 95)
print(f'El percentil 95 de la magnitud de los terremotos en Japón es: {perc}')
```

```python
# Encuentre el porcentaje de terremotos en Indonesia que se acompañaron de tsunamis
data2 = data[data.parsed_place == 'Indonesia']
totalTerremotos = data2.shape[0]
data2 = data[(data.parsed_place == 'Indonesia') & (data.tsunami == 1)]
terrConTsunami = data2.shape[0]
p = round((terrConTsunami/totalTerremotos)*100, 2)
print(f'Los terremotos acompañados de tsunamis en Indonesia son un {p}% del total')
```

```python
# - Calcule estadísticas resumidas para terremotos en Nevada
data3 = data[data.parsed_place == 'Nevada']
data3.describe()
```

```python
# - Agregue una columna que indique si el terremoto ocurrió en un país o estado de EE. UU. Que se encuentra en el "Anillo de fuego". - Utilice Alaska, Antarctica (buscar por Antarctic), Bolivia, California, Canada, Chile, Costa Rica, Ecuador, Fiji, Guatemala, Indonesia, Japan, Kermadec Islands, Mexico (no es New Mexico), New Zealand, Peru, Philippines, Russia, Taiwan, Tonga, and Washington.
anilloUSA = ['Alaska', 'California', 'Washington']
data4 = data.copy()
list = []
for e in data4['parsed_place']:
    if e in anilloUSA:
        list.append(True)
    else:
        list.append(False)
data4 = data4.assign(anillo_usa = list)
data4[['parsed_place', 'anillo_usa']].sample(30)
# data4[data4['parsed_place'] == 'Mexico']
```

```python
# - Calcula el número de terremotos en las ubicaciones del Anillo de Fuego y el número fuera de ellos
anillo = ['Alaska', 'Antarctic', 'Bolivia', 'California', 'Canada', 'Chile', 'Costa Rica', 'Ecuador', 'Fiji', 'Guatemala', 'Indonesia', 'Japan', 'Kermadec Islands', 'Mexico', 'New Zealand', 'Peru', 'Philippines', 'Russia', 'Taiwan', 'Tonga', 'Washington']
a = 0
b = 0
for e in data['parsed_place']:
    if e in anillo:
        a += 1
    else:
        b += 1
print(f'Los terremotos dentro del anillo de fuego son {a}. Los terremotos en el resto del mundo son {b}')
```

```python
# - Encuentra el recuento de tsunamis a lo largo del Anillo de fuego.
t = 0
for e in data['parsed_place']:
    if e in anillo:
        if data['tsunami'].bool == 1:
            t += 1

print(t)
```

```python
# - Encuentra el recuento de tsunamis a lo largo del Anillo de fuego.
t = 0
for n in range(data.shape[0]):
    if data['parsed_place'][n] in anillo:
        if data['tsunami'][n] == 1:
            t += 1

print(t)
```
