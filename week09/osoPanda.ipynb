```python
cols = ['time','title','place','magType','mag','alert','type','tsunami']
df = pd.read_csv('https://raw.githubusercontent.com/marcusRB/IDbootcamps_DataScience_student_PT_10201/master/dataset/earthquakes.csv', usecols=cols)
df.sample(10, random_state=17)
```

```python
df['nuevo'] = df.mag < 0
df.sample(20)
```

```python
# Operamos con expresiones regulares
df['place'].str.extract(r', (.*$)')[0].sort_values().unique()
```

```python
# Operamos con transformación de unos valores
df['parsed_place'] = df['place'].str.replace(
    r'.* of ', '', regex=True # elimina cualquier término que tenga of
).str.replace(
    'the ', '' # eliminamos "the "
).str.replace(
    r'CA$', 'California', regex=True # sustituye CA por California
).str.replace(
    r'NV$', 'Nevada', regex=True # sustituye NV por Nevada
).str.replace(
    r'MX$', 'Mexico', regex=True # sustituye MX por Mexico
).str.replace(
    r' region$', '', regex=True # elimina el término region
).str.replace(
    'northen ', '' # eliminamos 'northen ' 
).str.replace(
    'Fiji Islands', 'Fiji' # renombramos Fiji Island por Fiji
).str.replace(
    r'^.*, ', '', regex=True # elimina cualquier elemento raro desde el principio
).str.replace(
    r'Fiji Islan[a-zA-Z0-9]{1,}', 'Fiji', regex=True
).str.strip() # elimina espacios
df
```

```python
# Podemos hacer uso de assign() para crear columnas
df = df.assign(
    in_ca=df['parsed_place'].str.endswith('California'),
    in_alaska=df['parsed_place'].str.endswith('Alaska'),
    no = (df['in_ca'] == False) & (df['in_alaska'] == False)
).sample(20, random_state=20)
df[['parsed_place', 'in_ca', 'in_alaska', 'no']]
```
